```python
from dedalus2.public import *
```

```python
nx = 64
nz = 64
```

```python
Lz = 2*np.pi
```

```python
x_basis = Chebyshev(nx)
domain = Domain([x_basis],grid_dtype=np.complex128)
```

Here, we use `reshape()` to change the 1-D `z` array to a 2-D array.

```python
z = np.linspace(0,Lz,nz,endpoint=False)
zz = z.reshape(nz,1)
```

```python
figsize(12,8)
```

Note you only need to do this for the `z` array, not the `x` grid from Dedalus to make a 2D array by separation of variables.

```python
data_2D = np.cos(np.pi*domain.grid(0))*np.sin(zz)
```

```python
pcolormesh(domain.grid(0),z,data_2D)
```
