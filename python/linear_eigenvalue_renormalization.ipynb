```python
from dedalus2.public import *
from dedalus2.pde.solvers import LinearEigenvalue
```

```python
string = ParsedProblem(['x'], field_names=['f', 'fx'], param_names=['lam'])
xx_basis = Chebyshev(64, interval=(0.,1.))
str_domain = Domain([xx_basis])
```

```python
string.add_equation("dx(fx) + 1j*dt(f) = 0")
string.add_equation("dx(f) - fx = 0")
string.add_left_bc("f = 0")
string.add_right_bc("f = 0")
```

```python
string.expand(str_domain)
```

```python
string_lev = LinearEigenvalue(string,str_domain)
```

```python
string_lev.solve(string_lev.pencils[0])
```

```python
evals = string_lev.eigenvalues
```

```python
plot(string_lev.eigenvectors[:,3].real
     )
```

```python
indx = np.arange(len(evals))
e0 = indx[np.abs(evals) == np.nanmin(np.abs(evals))]
indices = np.argsort(evals)
```

```python
val = evals[e0]
```

```python
sqrt(val)
```
